```python
# Import the dependencies.
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 3. Create a new set of 2,000 random latitudes and longitudes.
lats = np.random.uniform(low=-90.000, high=90.000, size=2000)
lngs = np.random.uniform(low=-180.000, high=180.000, size=2000)
lat_lngs = zip(lats, lngs)
lat_lngs
```

```python
# Add the latitudes and longitudes to a list.
coordinates = list(lat_lngs)
```

```python
# 4. Get the nearest city using the citipy module.

# Self-note: generally this line from citipy import citipy should be in the above cell (Import the Dependencies),
# however, I ran into a problem: the code did not work when I initially created this file (Weather_Database.ipynb)
# with using Python 3 (ipykernel), instead this worked when creating the file with PythonData (not Python 3 (ipykernel)).
# For future projects, it will look neater/more organized to follow if placed above.

from citipy import citipy 
```

```python
# 4. Get the nearest city using the citipy module.

# Create a list for holding the cities.
cities = []
# Identify the nearest city for each latitude and longitude combination.
for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name

    # If the city is unique, then we will add it to the cities list.
    if city not in cities:
        cities.append(city)
        
# Print the city count to confirm sufficient count.
len(cities)
```
